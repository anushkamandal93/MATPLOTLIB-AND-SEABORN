```python
#SAMPLE DATA
X=[1,2,3,4,5]
Y=[2,4,6,8,10]
```

```python
#INSET AXES

fig=plt.figure(figsize=(6,4))
axes1=fig.add_axes([0,0,1,1]) #L,B,H,W
axes1.plot(X,Y)
plt.title("the axis")

axes2=fig.add_axes([0.2,0.5,0.2,0.2])
axes2.plot(Y,X)
plt.title("the graph")


plt.show()
```

```python
#DPI (zoom in-zoom out)

x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.figure(figsize=(6,4),dpi=100)
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[1,3,5,7,9]

plt.plot(x,y1,label='blue')
plt.plot(x,y2,label='red')
plt.title('title')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

```

```python
#LINES

x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[1,3,5,7,9]

plt.plot(x,y1,label='blue',lw=3.5,ls=':')
plt.plot(x,y2,label='red',ls='-.')
```

```python
#MARKERS (helps to decide co-ordinates & you can use any sign)

x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[1,3,5,7,9]

plt.plot(x,y1,label='blue',lw=3.5,ls=':',marker='o')
plt.plot(x,y2,label='red',ls='-.',marker='*')

```

**SCATTER,HISTOGRAM & BOXPLOT**

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[5,7,4,6,8]

#SCATTER:- r/ship b/w two variables
plt.scatter(x,y)
```

```python
#HISTOGRAM:- tells shape and spread
import numpy as np
nums = np.random.randn(100)
plt.hist(nums,bins=20) #bins for bars gap
```

```python
#BOXPLOT:- single col. ki outliers btata hai

#5 summary points:-
#min,
#25% --> lower quatile
#50% --> median
#75% --> upper quatile
#max

''' values b/w lower and upper quatile are called generic or actual values
plt.boxplot(nums)

value outside of lower and upper quatile are called outlier '''
```
